```python
import os, glob, shutil, pandas as pd

# Find the CSV (Kaggle uses "WA_Fn-UseC_-Telco-Customer-Churn.csv")
candidates = [p for p in glob.glob("data/*.csv") if "Telco" in os.path.basename(p)]
if not candidates:
    # fallback: any single csv in data/
    candidates = glob.glob("data/*.csv")

assert len(candidates) >= 1, "CSV not found in data/ after unzip."

src = candidates[0]
dst = "data/telco_raw.csv"
if src != dst:
    shutil.move(src, dst)

print("Raw CSV at:", dst)

# lightweight load test
df = pd.read_csv(dst)
print("Rows, Cols:", df.shape)
print("Columns:", list(df.columns)[:8], "...")
df.head(3)
```
